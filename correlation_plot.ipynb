```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
np.random.seed(31415)
```

```python
data = pd.read_csv('/Users/bradyhong/Downloads/SoccerPredictions-main-4/Data/Premier_league_10_20.csv')
```

```python
def assign(row):
    if row.home_score > row.away_score:
        return 'H'
    elif row.home_score < row.away_score:
        return 'A'
    else:
        return 'D'
data['result'] = data.apply (lambda row: assign(row), axis=1)
```

```python
x = data.iloc[:,3:5]
x1 = data.iloc[:,23:27]
x2 = data.iloc[:,33:39]
x_new = pd.concat([x, x1], axis = 1)
x_new = pd.concat([x_new, x2], axis = 1)
y = data.iloc[:,-1]
d_new = pd.concat([x_new, y], axis = 1)
```

```python
import seaborn as sns
#get correlations of each features in dataset
corrmat = d_new.corr()
top_corr_features = corrmat.index
plt.figure(figsize=(20,20))
#plot heat map
sns.set(font_scale=2)
g=sns.heatmap(data[top_corr_features].corr(),annot=True,cmap = plt.get_cmap('vlag'),center=0,fmt="0.2f", linewidths=.5)

g.axes.set_title("Correlation plot of the different soccer in-game statistics",fontsize=40)
g.tick_params(labelsize=30)
```
